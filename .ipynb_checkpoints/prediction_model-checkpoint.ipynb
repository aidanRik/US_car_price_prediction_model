```python
#Splitting the data and Target data 
```

```python
data = dataset.drop(["Car_Name", "Selling_Price"], axis = 1)
target_data = dataset["Selling_Price"]
```

```python
print(data)
print(target_data)
```

```python
#Splitting the train and test data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data, target_data, test_size = 0.1, random_state = 2)
```

```python
#Model training
```

```python
#1. linear regression
```

```python
lin_reg_model = LinearRegression()
```

lin_reg_model.fit(X_train, Y_train)

```python
# Evaluation on training
training_data_predict = lin_reg_model.predict(X_train)
```

```python
# r score error 
error = metrics.r2_score(Y_train, training_data_predict)
print("r score error:", error)
```

```python
#visualize
plt.scatter(Y_train, training_data_predict)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual prices vs predicted prices")
plt.show
```

```python
# evaluation on test data
test_data_predict = lin_reg_model.predict(X_test)
```

```python
# r score error 
error = metrics.r2_score(Y_test, test_data_predict)
print("r score error:", error)
```

```python
#visualize
plt.scatter(Y_test, test_data_predict)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual prices vs predicted prices")
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
big_dataset = pd.read_csv("../car_price_prediction_model/CAR DETAILS FROM CAR DEKHO.csv")
```

```python
big_dataset.head(10)
```

```python
print(big_dataset.fuel.value_counts())
print(big_dataset.seller_type.value_counts())
print(big_dataset.transmission.value_counts())
print(big_dataset.owner.value_counts())
```

```python
#encode the categorical data
big_dataset.replace({"seller_type":{"Dealer":0, "Individual":1, "Trustmark Dealer":2}}, inplace = True)
big_dataset.replace({"transmission":{"Manual":0, "Automatic":1}}, inplace = True)
big_dataset.replace({"fuel":{"Petrol":0, "Diesel":1, "CNG":2, "LPG":3, "Electric":4}}, inplace = True)
big_dataset.replace({"owner":{"First Owner":0, "Second Owner":1, "Third Owner":2, "Fourth & Above Owner":3, "Test Drive Car":4}}, inplace = True)
big_dataset.head()
```

```python
# Split-out validation dataset
x = big_dataset.drop(["name", "selling_price"], axis = 1)
y = big_dataset["selling_price"]
X_train, X_validation, Y_train, Y_validation = train_test_split(x, y, test_size=0.20, random_state=1, shuffle=True)
```

```python
# load model and fit
lasso_model = Lasso()
lasso_model.fit(X_train, Y_train)
```

```python
#Evaluation
train_data_predict = lasso_model.predict(X_train)
#r score
error = metrics.r2_score(Y_train, train_data_predict)
print("r score error:", error)
```

```python
plt.scatter(Y_train, train_data_predict)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual prices vs predicted prices")
plt.show
```
